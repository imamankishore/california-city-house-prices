```python
df = pd.read_csv('housing.csv')
```

```python
df['total_bedrooms'] = df['total_bedrooms'].fillna(df['total_bedrooms'].mean())
```

```python
df['rooms_per_household'] = np.round((df['total_rooms'] + df['total_bedrooms'])/df['households'])
```

```python
df = df.drop(columns=['total_rooms', 'total_bedrooms' , 'households'])
```

```python
# Define which columns to apply each transformer to
categorical_cols = ['ocean_proximity']
numerical_cols = ['longitude', 'latitude', 'housing_median_age', 'population','median_income' ,'rooms_per_household']

# Create the transformer
trf = ColumnTransformer([
    ('ohe', OneHotEncoder(drop='first'),[-2]),
    ('scaler', MinMaxScaler(feature_range=(0, 1)), slice(None, -2))
], remainder='passthrough')
```

```python
trf3 = RandomForestRegressor()
```

```python
pipe = make_pipeline(trf,trf3)
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred = pipe.predict(x_test.values)
```
